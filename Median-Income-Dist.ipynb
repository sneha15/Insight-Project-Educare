```python
students = pd.read_pickle('students')
students.head(800)
```

```python
#load census data

income1 = pd.read_csv('/home/harisk87/Dropbox/PythonProjects/Insight-Project/IncomeData.csv')
income = income1[['GEO.id2', 'HC02_EST_VC02', 'HC02_EST_VC04']]
income = income.rename(columns={'GEO.id2':'zip_student'})
#income.head()
# get zip codes and median income
students = pd.merge(students, income, on='zip_student')
# put in median income into clumns
students = students.rename(columns={'HC02_EST_VC02': 'household_income', 'HC02_EST_VC04': 'median_income'})

```

```python
# clean median and household income

# plot distribution of household income
trace1 = Histogram(x=students['household_income'])
data = [trace1]
layout = Layout(xaxis=dict(title = 'Household Income ($)'),yaxis=dict(title='Number of students'),title='Household Income Distribution')


fig1 = Figure(data=data, layout=layout)
iplot(fig1)

```

```python
# plot distribution of median income
trace1 = Histogram(x=students['median_income'])
data = [trace1]
layout = Layout(xaxis=dict(title = 'Median Income ($)'),yaxis=dict(title='Number of students'),title='Median Income Distribution')


fig1 = Figure(data=data, layout=layout)
iplot(fig1)
```
